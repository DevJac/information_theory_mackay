```julia
using Plots
using StatsBase
```

*Channel capacity* is the *rate* at which information can be reliably transmitted over a communication channel.

$$C(f) = 1 - H_2(f)$$

```julia
plot(0:0.01:0.5, f -> 1 - entropy([f, 1-f], 2), legend=false)
```

Thus, the more noisy a channel is, the lower its *rate* of reliable communication.

A *rate* of 0.5 would mean that our coding system is only able to reliably transmit information half as quickly as it could when transmitting raw data on a perfect channel.

An interesting observation is that when $f = 0.5$ no reliable communication is possible at any speed.
